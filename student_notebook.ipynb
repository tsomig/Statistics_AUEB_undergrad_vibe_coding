## 💾 Παραγωγή του `submission.json`

Εκτελέστε το παρακάτω κελί ΜΟΝΟ αφού έχετε συμπληρώσει όλες τις απαντήσεις παραπάνω.

Στο Colab, το αρχείο εμφανίζεται στο αριστερό panel (📁 Files) και μπορείτε να το κατεβάσετε με δεξί-κλικ → Download.

```python
import json

# Συνάρτηση που μετατρέπει None σε None στο JSON, αλλιώς float/int
def opt_float(x):
    return None if x is None else float(x)

def opt_int(x):
    return None if x is None else int(x)

submission = {
    "AM": AM,
    "answers": {
        # --- CORE (υποχρεωτικά) ---
        "q1_mean_X":             float(q1_mean_X),
        "q1_mean_Y":             float(q1_mean_Y),
        "q1_var_Y":              float(q1_var_Y),
        "q1_cdf_X_at_tau":       float(q1_cdf_X_at_tau),
        "q2_chebyshev_bound":    float(q2_chebyshev_bound),
        "q2_empirical_frac":     float(q2_empirical_frac),
        "q3_posterior":          float(q3_posterior),
        "q4_cov_YW":             float(q4_cov_YW),
        "q4_rho_YW":             float(q4_rho_YW),
        "q4_slope":              float(q4_slope),
        "q5_cond_mean":          float(q5_cond_mean),
        "q5_cond_sd":            float(q5_cond_sd),
        "q5_audit_threshold_w0": float(q5_audit_threshold_w0),
        # --- BONUS (προαιρετικά — None αν παρακάμφθηκαν) ---
        "q5_emp_quantile_residuals": opt_float(q5_emp_quantile_residuals),
        "q5_n_flagged_model":        opt_int(q5_n_flagged_model),
        "q5_n_flagged_empirical":    opt_int(q5_n_flagged_empirical),
    }
}

out_path = f"{AM}_submission.json"
with open(out_path, "w", encoding="utf-8") as f:
    json.dump(submission, f, indent=2, ensure_ascii=False)

n_bonus_done = sum(1 for v in [submission['answers']['q5_emp_quantile_residuals'],
                                submission['answers']['q5_n_flagged_model'],
                                submission['answers']['q5_n_flagged_empirical']]
                   if v is not None)
print(f"✅ Αποθηκεύτηκε: {out_path}")
print(f"   Bonus που συμπληρώθηκαν: {n_bonus_done}/3")
print()
print(json.dumps(submission, indent=2, ensure_ascii=False))
```
